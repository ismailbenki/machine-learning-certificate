```python
import numpy as np
import matplotlib.pyplot as plt
from display import gaussian_samples

data,labels = gaussian_samples(100,N=4,show=True)
```

```python
from sklearn import svm
from display import display_classifier

svm_clf = svm.SVC(kernel='poly',degree=3,coef0=0)
svm_clf.fit(data,labels)
    
display_classifier(svm_clf,data,labels)
```

```python
svm_clf = svm.SVC(kernel='rbf',gamma=2)
svm_clf.fit(data,labels)

display_classifier(svm_clf,data,labels)
```

```python
svm_clf = svm.SVC(kernel='linear')
svm_clf.fit(data,labels)

display_classifier(svm_clf,data,labels)
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

iris = datasets.load_iris()
iris_data = iris.data[:,:2]
iris_train, iris_test, iris_y_train, iris_y_test = train_test_split(iris_data,
                                                                iris.target,
                                                                test_size=0.2,
                                                                random_state=0)
```

```python
svm_clf = svm.SVC(kernel='poly',degree=3,coef0=.7)
svm_clf.fit(iris_train,iris_y_train)

display_classifier(svm_clf,iris_data,iris.target)

predicted = svm_clf.predict(iris_test)
print('Matrice de confusion :\n',confusion_matrix(iris_y_test,predicted))
print(classification_report(iris_y_test,predicted))                                    
```
